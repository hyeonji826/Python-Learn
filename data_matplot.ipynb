## 기본 그래프 그리기
### 그래프의 종류
- 선 그래프 : 시간에 따른 변화나 연속적인 데이터를 표현할 때 사용한다.
- 막대 그래프 : 범주별 비교나 빈도, 양을 비교할 때 효과적이다.
- 산점도(scatter point): 두 변수 간의 관계를 보여줄 때 유용하다.
- 파이 차트(pie chart) : 전체에 대한 부분의 비율을 표현할 때 사용
- 히스토그램(histogram) : 연속적인 데이터의 분포를 나타낼 때 사용

### 선 그래프
''' python
plt.ploy(x,y)
'''
- 위 코드로 선 그래프를 그릴 수 있다.

````python
# matplotlib과 numpy 라이브러리 임포트
import matplotlib.pyplot as plt
import numpy as np
````

````python
# x축 데이터 생성(0부터10까지 100갸의 균등한 간격의 점 생성)
x = np.linspace(0,10,100)
print(x)

# y축 데이터 생성(sin함수 적용)
y =np.sin(x)
print(y)
````

````python
# 그래프 그리기
plt.plot(x,y)

# 그래프 상단에 타이틀 정보 입력(기본 한글 깨짐)
plt.title("Title")

# 그래프 표시
plt.show()
````

````python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[10,20,25,30,40]

plt.plot(x,y,marker='0',linestyle='-',color='b',label="데이터")
plt.xlabel("X축")
plt.ylabel("Y축")
plt.title("Basic line plot")
plt.grid(True)
plt.show()
````

````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y, marker='o', linestyle='-', color='b', label="데이터")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("기본 선 그래프")
plt.legend()
plt.grid(True)
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(100)		# 정규 분포를 따르는 1000개의 난수 생성
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("histogram practice")
plt.show()
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 랜덤 데이터 생성
np.random.seed(42)
data = np.random.randn(1000)

# 히스토그램과 커널 밀도 그래프(KDE)
sns.histplot(data, kde=True, color="blue")
plt.title("Seaborn histogram practice")
plt.show()
````

### 막대 그래프
barplot
- 범주형 데이터를 시각화하는 데 효과적인 방법이다.
- 각 범주(catrgory)는 높이나 길이가 데이터 값에 비례하는 막대로 표현된다.

#### 그래프의 특징
- 데이터 비교 : 서로 다른 범주 간에 수량이나 크기, 길이 등을 쉽게 비교할 수 있다.
- 간단한 해석 : 직관적이어서 쉽게 이해할 수 있다.
- 다양한 형태 : 세로 막대가 기본이지만 가로 막대나 누적 막대 등 다양한 형태로 표현될 수 있다.

'''python
df = pd.DataFrame(data)
plt.bar(df['x축'],df['y축'])
'''

위 코드로 막대 그래프를 그릴 수 있다.

````python
# 데이터 프레임을 생성하기 위한 pandas 임포트
import pandas as pd
````

````python
# 더미 데이터 생성
# 각 value의 리스트 길이는 모두 동일해야 한다.
data ={
    'fruit':['apple','banana','orange'],
    'sell_amount':[45,32,50]
}
# 데이터프레임으로 변환
df=pd.DataFrame(data)
````

````python
# 막대 그래프로 시각화
# 첫번째 인수는 x축에 배치, 두번째 인수는 y축에 배치
plt.bar(df['fruit'],df['sell_amount'])

# 시각화
plt.show()
````

````python
# seaborn 라이브버리 임포트
import seaborn as sns

# 예제 데이터셋 로드
tips = sns.load_dataset("tips")
print(tips)
````

````python
# 산점도 그리기
# x축은 총 계산서
# y축은 팁
# 두 변수의 관계를 산점도로 표현한다.
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.show()
````
